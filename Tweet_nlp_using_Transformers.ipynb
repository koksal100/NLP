```python
oğuz["Tweet"]=oğuz["Tweet"].apply(lambda x:x.lower())
```

```python
def combine(ner_result):
    combined_entities = []
    current_entity =""
    current_word=""
    b_counter=0
    for entity in ner_result:
        if entity['entity'][0]=="B":
          if b_counter==1:
            combined_entities.append({"entity":current_entity,"word":current_word})
            current_entity = entity["entity"][2:]
            current_word=entity["word"]
            continue
          b_counter=1
          current_entity = entity["entity"][2:]
          current_word=entity["word"]
        elif entity['entity'][0]=="I":
          if entity["word"][:2]=="##":
            entity["word"]=entity["word"][2:]
            current_word=current_word+entity["word"]
            continue
          current_word=current_word+" "+entity["word"]


    if current_entity is not None:
        combined_entities.append({"entity":current_entity,"word":current_word})

    return combined_entities

```

```python
combine(pipe2(oğuz["Tweet"][0]))
```

```python
oğuz["ner"]=oğuz["Tweet"].apply(lambda x: combine(pipe2(x)))
```

```python
oğuz["ner"]
```

```python
persons=[]

for i in range(len(oğuz)):
  for j in range(len(oğuz["ner"][i])):
    if oğuz["ner"][i][j]["entity"]=="PER":
      persons.append(oğuz["ner"][i][j]["word"])
```

```python
organizations=[]

for i in range(len(oğuz)):
  for j in range(len(oğuz["ner"][i])):
    if oğuz["ner"][i][j]["entity"]=="ORG":
      organizations.append(oğuz["ner"][i][j]["word"])
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Örnek liste

# Elemanların frekansını hesapla
eleman_frekans = Counter(organizations)

# Elemanları ve frekanslarını ayır
elemanlar = list(eleman_frekans.keys())
frekanslar = list(eleman_frekans.values())

# Grafik oluştur
plt.figure(figsize=(10, 6))
plt.bar(elemanlar, frekanslar, color='skyblue')
plt.xlabel('Elemanlar')
plt.ylabel('Frekans')
plt.title('Liste Elemanlarının Dağılımı')
plt.xticks(rotation=90)
plt.show()
```

```python
from googletrans import Translator
```

```python

# Çeviri yapıcıyı oluştur
translator = Translator()

# Çeviri fonksiyonu
def translate_text(text):
    try:
        translation = translator.translate(text, src='tr', dest='en')
        return translation.text
    except Exception as e:
        return f"Error: {e}"

# 'text' sütununu İngilizce'ye çevir ve yeni bir sütun olarak ekle
oğuz['Tweet_en'] = oğuz['Tweet'].apply(translate_text)

# Sonucu göster
print(oğuz)
```
